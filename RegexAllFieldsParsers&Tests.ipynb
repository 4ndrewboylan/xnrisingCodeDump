## MLA

prototype:

Author(s). "Title of Article." Title of Journal, Volume, Issue, Year, pages.

examples:

Bagchi, Alaknanda. "Conflicting Nationalisms: The Voice of the Subaltern in Mahasweta Devi's Bashai Tudu." Tulsa Studies in Women's Literature, vol. 15, no. 1, 1996, pp. 41-50.

Duvall, John N. "The (Super)Marketplace of Images: Television as Unmediated Mediation in DeLillo's White Noise." Arizona Quarterly, vol. 50, no. 3, 1994, pp. 127-53.

Approach 1 (Previous Strategy): mla regex for matching the whole raw ref. Not an ideal approach:

```python
mla_regex = r"(?P<num>[0-9\[\]]+\.?)?\s?(?P<authors>[\w\,\.\s]+.)[\s]?[\"\“](?P<title>[\w –\'\-\s\:.,\-\(\)]+[.]?)[\"\”]\s?(?P<publication>[\w\-\s\:\,\-\(\)]+)[,]?[\s]?(?P<volume>(vol.)?\s?[\w0-9.,\(\)]+)\s?(?P<year>[\(]\d{4}[\)])[\:\,]?(?P<index>(pp.)?\s[0-9–\-.,]+)"
```

Approach 2 (Current Strategy): use mla regex for individual fields. Work well:

```python

def testMlaFields() -> list:
    all_raw_strings = []
    # titles collections
    all_truth_titles = []
    all_parsed_titles = []
    titles_match = []
    # authors collections
    all_truth_authors = []
    all_parsed_authors = []
    authors_match = []
    # volumes collections
    all_truth_volumes = []
    all_parsed_volumes = []
    volumes_match = []
    # issues collections
    all_truth_issues = []
    all_parsed_issues = []
    issues_match = []
    # years collections
    all_truth_years = []
    all_parsed_years = []
    years_match = []
    # publications collections
    all_truth_publications = []
    all_parsed_publications = []
    publications_match = []
    # pages collections
    all_truth_pages = []
    all_parsed_pages = []
    pages_match = []


    # matched fields counters
    titles_match_count = 0        
    authors_match_count = 0 
    volumes_match_count = 0
    issues_match_count = 0    
    years_match_count = 0
    publications_match_count = 0        
    pages_match_count = 0        
    

    # Parse all matching ones from the whole 30k raw refs
    for index, row in original_df.iterrows():
        ref = row['mla']
        # # Using corresponding style parser
        # result = re.search(mla_regex, ref)  

        all_raw_strings.append(ref)
        # Check title
        all_truth_titles.append(row['Title'])
        if '[' in ref and ']' in ref and ('“' in ref or '"' in ref):
          if ('“' in ref and (ref.index(']') < ref.index('”')
          or ref.index('[') - ref.index('”') > 2)):
            # print(ref)
            title = re.search('“(.+?)”', ref).group(1) 
          # elif ('"' in ref and (ref.index(']') < ref.index('"')
          # or ref.index('[') - ref.index('"') > 2)):
          #   title = re.search('"(.+?)"', ref).group(1)  
          else:
            title = re.search('\[(.+?)\]', ref).group(1)
        elif '“' in ref:
          title = re.search('“(.+?)”', ref).group(1) 
        # elif '"' in ref:
        #   # print(ref)
        #   title = re.search('"(.+?)"', ref).group(1)  
        else:
          title = ""
        all_parsed_titles.append(title)
        # Strip punctuations wrapping around title truth
        if (row['Title'].strip("[].") == title.strip(".")):
            titles_match.append(1)
            titles_match_count += 1
        else:
            titles_match.append(0)

        # Check author
        truth_author_arr = row['Author']
        all_truth_authors.append(truth_author_arr)
        a = []
        if ('"' in ref):
          a = ref.split('"')
        if ('et al' in ref):
          a = ref.split('et al')
        else:
          a = ref.split('“')
        author = a[0]
        temp = []
        if ' and ' in author:
          temp = author.split(' and ')
          author = "".join(temp)
        if ' Jr' in author:
          temp = author.split(' Jr')
          author = "".join(temp)
        if ' 2nd' in author:
          temp = author.split(' 2nd')
          author = "".join(temp)
        if ' 3rd' in author:
          temp = author.split(' 3rd')
          author = "".join(temp)
        if ' Sr' in author:
          temp = author.split(' Sr')
          author = "".join(temp)
        all_parsed_authors.append(author)

        # Use counters to see if truth and parsed authors matches
        author_match = True
        parsed_author_string = ""
        for c in author:
            if c.isalnum():
                special_char = "[]&.,-"
                if (c not in special_char) and (c != " "):
                    parsed_author_string += c

        # Strip all special symbols and spaces for both author group string in 
        # ground truth and the parsed authors
        truth_author_string = ""
        for word in truth_author_arr:
          for c in word:
            if c.isalnum():
              special_char = "[]&.,-"
              if (c not in special_char) and (c != " "):
                  truth_author_string += c

        length = len(parsed_author_string)
        truth_author_string = truth_author_string[0:length]

        if Counter(parsed_author_string) != Counter(truth_author_string):
          author_match = False
          # print(author, parsed_author_string, truth_author_string)

        if author_match:
            authors_match.append(1)
            authors_match_count += 1
        else:
            authors_match.append(0)

        #Check volume, issue, and year
        volume = ""
        issue = ""
        year = ""
        r = re.search('((vol. )(?P<vol>[0-9]+)\s?[,]?)(?P<issue>[\w0-9\-. ]+)[,]?\s?([\(]?(?P<year>\d{4})[\)]?[\:\,]?)', ref)
        if r:
          volume = r.group('vol')
          issue = r.group('issue')
        if 'vol.' not in ref:
          issue = re.search('(\s?(?P<issue>[\w0-9\-. ]+)[,]?\s?([\(]?(?P<year>\d{4})[\)]?[\:\,]+))', ref).group('issue')

        # Check volumes
        if 'vol.' in ref:
          # print(ref)
          v = re.search('((vol.)\s?[0-9]+[, ])', ref)
          if v:
            volume = v.group(1)
            volume = volume.strip(',')
            volume = volume.strip('vol.')
            volume = volume.strip()
          v = re.search('((vol.)\s?[0-9]+[, ][A-Za-z]+\s)', ref)
          if v:
            volume = v.group(1)
            volume = volume.strip('vol.')
            volume = volume.strip()

        all_parsed_volumes.append(volume)            
        all_truth_volumes.append(row['Volume'])

        if (row['Volume'].strip() == volume):
            volumes_match.append(1)
            volumes_match_count += 1
        elif (row['Volume'].strip() == 'VolumeNotFound') and (volume == ""):
            volumes_match.append(1)
            volumes_match_count += 1
        else:
            volumes_match.append(0)

        # Check issues
        all_parsed_issues.append(issue)            
        all_truth_issues.append(row['Issue'])

        if (str(row['Issue']).strip() == str(issue).strip()):
            issues_match.append(1)
            issues_match_count += 1
        elif (str(row['Issue']).strip() == 'IssueNotFound') and (str(issue).strip() == ''):
            issues_match.append(1)
            issues_match_count += 1
        else:
            issues_match.append(0)

        # Check years
        # Based on the current ground truth, year is wrapped with () but mla style does not require ()
        # Currently set the () a prerequisite for year field
        y = re.search('([\(](?P<year>\d{4})[\)][\:\,]+)', ref)
        # Check if the raw ref contains multiple years 
        # e.g. (1946) vol. 101,3 (1976): 88-90. doi:10.1055/s-0028-1104041
        multipleY = re.search('([\(]?(\d{4})[\)]?)(.+?)([\(](?P<year>\d{4})[\)][\:\,]+)', ref)
        if y and (not multipleY):
          year = y.group('year').strip()
        elif multipleY:
          year = multipleY.group('year').strip()
        all_parsed_years.append(year)
        all_truth_years.append(row['Year'])

        if (str(row['Year']) == year):
            years_match.append(1)
            years_match_count += 1
        elif (str(row['Year']) == 'YearNotFound') and (year == ""):
            years_match.append(1)
            years_match_count += 1
        else:
            years_match.append(0)

        # Check publications
        pub = ''
        re_title_pub_vol = '[\"\“](?P<title>[\w –\'\-\s\:.,\-\(\)]+[.]?)[\"\”]\s?(?P<publication>[\w\-\'\s\:\,\-\(\)]+)[,]?[\s]?(?P<volume>(vol.)\s?[\w0-9.,\(\)]+)\s?'
        r_title_pub_vol = re.search(re_title_pub_vol, ref)
        re_pub_one = '(.\”)\s(?P<publication>[\w., ]+)\s(vol.)\s*'
        r_pub_one = re.search(re_pub_one, ref)
        re_pub_two = '(.\”)\s(?P<publication>[\w.:\-\;\' ]+)\s[(vol.)(,)]\s*'
        r_pub_two = re.search(re_pub_two, ref)
        if r_title_pub_vol:
            pub = r_title_pub_vol.group('publication').strip(',')
            pub = pub.strip()
        elif r_pub_one:
            pub = r_pub_one.group('publication').strip()   
        elif r_pub_two:
            pub = r_pub_two.group('publication').strip()

        # Check if the journal name is between '].' and 'vol.'
        r_foreign_pub = re.search('(].)\s(?P<publication>[\w\'\-\:\. ]+)\s[(vol.),]\s*', ref)
        if r_foreign_pub: 
            pub = r_foreign_pub.group('publication').strip()

        all_truth_publications.append(row['JournalName'])
        all_parsed_publications.append(pub);
        # Strip punctuations wrapping around title truth
        if (row['JournalName'].strip() == pub):
            publications_match.append(1)
            publications_match_count += 1
        else:
            publications_match.append(0)

        
        # Check pages
        page = ''
        re_page_one = '(?P<year>[\(]?\d{4}[\)]?)[\:\,]?\s(?P<page>(pp.)?\s?[\w\0-9]+)(.)[\n]\s*'
        re_page_two = '(?P<year>[\(]?\d{4}[\)]?)[\:\,]?\s(?P<page>(pp.)?\s?[0-9A-Z\–\-\, ]+)(.)\s*'
        r_page_one = re.search(re_page_one, ref)
        r_page_two = re.search(re_page_two, ref)
        if r_page_one:
            page = r_page_one.group('page')
            page.strip()
        elif r_page_two:
            page = r_page_two.group('page')
            page.strip()
        all_parsed_pages.append(page)            
        all_truth_pages.append(row['PageNumber'])

        if (row['PageNumber'].strip() == page):
            pages_match.append(1)
            pages_match_count += 1
        elif (row['PageNumber'].strip() == 'UNKNOWN') and (page == ''):
            pages_match.append(1)
            pages_match_count += 1
        else:
            pages_match.append(0)



    components = {
        'Raw' : all_raw_strings,
        'Title Truth': all_truth_titles,
        'Title Parsed': all_parsed_titles,
        'Title Match': titles_match,
        'Author Truth': all_truth_authors,
        'Author Parsed': all_parsed_authors,
        'Author Match': authors_match,
        'Volume Truth': all_truth_volumes,
        'Volume Parsed': all_parsed_volumes,
        'Volume Match': volumes_match,
        'Issue Truth': all_truth_issues,
        'Issue Parsed': all_parsed_issues,
        'Issue Match': issues_match,
        'Year Truth': all_truth_years,
        'Year Parsed': all_parsed_years,
        'Year Match': years_match,
        'JournalName Truth': all_truth_publications,
        'JournalName Parsed': all_parsed_publications,
        'JournalName Match': publications_match,
        'PageNumber Truth': all_truth_pages,
        'PageNumber Parsed': all_parsed_pages,
        'PageNumber Match': pages_match
    }

    result = []
    result.append(components)
    result.append(titles_match_count)
    result.append(authors_match_count)
    result.append(volumes_match_count)
    result.append(issues_match_count)
    result.append(years_match_count)
    result.append(publications_match_count)
    result.append(pages_match_count)

    return result
```

```python
mla_result = testMlaFields()
mla_components = mla_result[0]
mla_title_match_count = mla_result[1]
mla_author_match_count = mla_result[2]
mla_volume_match_count = mla_result[3]
mla_issue_match_count = mla_result[4]
mla_year_match_count = mla_result[5]
mla_publication_match_count = mla_result[6]
mla_page_match_count = mla_result[7]
```

##APA
prototype:

Article Author, A. A., & Article Author, B. B. (Year). Title of article. Title of Journal, volume number(issue number), inclusive page numbers.

examples:

Hanna, K. (2007). Adsorption of aromatic carboxylate compounds on the surface of synthesized iron oxide-coated sands. Applied Geochemistry, 22(9), 2045-2053.

Kyratsis, A. (2004). Talk and interaction among children and the co-construction of peer groups and peer culture. Annual Review of Anthropology, 33(4), 231-247.

Parker, G., & Roy, K. (2001). Adolescent depression: A review. Australian and New Zealand Journal of Psychiatry, 35(5), 572-580.

Zhang, B., Cohen, J., Ferrence, R., & Rehm, J. (2006). The impact of tobacco tax cuts on smoking initiation among Canadian young adults. American Journal of Preventive Medicine, 30(6), 474-479.

reference: apa regexes for matching the whole raw ref. (Original failed approach)

```python
apa_regex_noissue = r"(?P<num>[\[\]0-9]+\.?)?\s?(?P<authors>(( and )?( & )?([\w]+)[,]?((\s*[\-]?[A-Z][. ]){1,2})[, ]?\s?(Jr )?)+)( et al. )?[\(]?(?P<year>(\d{4}))[\)]?[.]?[ ]?(?P<title>[A-Z][\w:.,;–\+\-\/\(\) ]+[.])(?P<publication>[\w. ]+)[.,]*\s(?P<volume>[\w0-9\(\)]+)[,]\s?(?P<index>[\w0-9\–]+)\s*"
apa_regex_withissue = r"(?P<num>[\[\]0-9]+\.?)?\s?(?P<authors>(( and )?( & )?([\w]+)[,]?((\s*[\-]?[A-Z][. ]){1,2})[, ]?\s?(Jr )?)+)( et al. )?[\(]?(?P<year>(\d{4}))[\)]?[.]?[ ]?(?P<title>[A-Z][\w:.,;–\+\-\/\(\) ]+[.])(?P<publication>[\w. ]+)[.,]*\s(?P<volume>[\w0-9]+)[\(](?P<issue>[0-9]+)[\)][,]\s?(?P<index>[0-9\–\-]+)\s*"
```

reference: apa regexes for matching text and nontext fields, implemented in the function below (New approach)

```python
apa_regex_justtext = r"(?P<authors>(.+?))[\(](?P<year>(\d{4}))[\)][.]?[ ]?(?P<title>[A-Za-z0-9][\w\:.,?;\%\/\–\-\'\"\+\(\)\[\] ]+[.])\s?(?P<publication>[\w\'\-\;:,. ]+)[,]\s[\(]?[0-9][\)]?\s*"
```

```python
apa_regex_nontext = r"[., ]+(?P<volume>[\0-9]+)([\(](?P<issue>[0-9]{1,2})[\)])?[,]\s?(?P<page>([0-9]+[A-Z]?[a-z]?)[\–\-]?([0-9]+[A-Z]?[a-z]?))\s*"
```

```python

def testApaFields() -> list:
    all_raw_strings = []
    # titles collections
    all_truth_titles = []
    all_parsed_titles = []
    titles_match = []
    # authors collections
    all_truth_authors = []
    all_parsed_authors = []
    authors_match = []
    # volumes collections
    all_truth_volumes = []
    all_parsed_volumes = []
    volumes_match = []
    # issues collections
    all_truth_issues = []
    all_parsed_issues = []
    issues_match = []
    # years collections
    all_truth_years = []
    all_parsed_years = []
    years_match = []
    # publications collections
    all_truth_publications = []
    all_parsed_publications = []
    publications_match = []
    # pages collections
    all_truth_pages = []
    all_parsed_pages = []
    pages_match = []

    # matched fields counters
    titles_match_count = 0        
    authors_match_count = 0 
    volumes_match_count = 0
    issues_match_count = 0    
    years_match_count = 0        
    publications_match_count = 0        
    pages_match_count = 0 


    # Parse all matching ones from the whole 30k raw refs
    for index, row in original_df.iterrows():
        ref = row['apa']
        author = ''
        year = ''
        title = ''
        volume = ''
        issue = ''
        pub = ''
        page = ''

        # apa regex contains text matching
        result_justtext = re.search(apa_regex_justtext, ref)
        # apa regex for nontext matching
        result_nontext = re.search(apa_regex_nontext, ref)

        # Match just text part
        if result_justtext:
            author = result_justtext.group('authors')
            year = result_justtext.group('year')
            title = result_justtext.group('title')
            pub = result_justtext.group('publication')
        # Match the nontext part: volume, issue
        if result_nontext:
            volume = result_nontext.group('volume')
            issue = result_nontext.group('issue')
            page = result_nontext.group('page')

          
        all_raw_strings.append(ref)

        # Check author
        a_re = re.search('(?P<authors>(.+?))[\(](?P<year>(\d{4}))[\)][.]?[ ]?(.+?)\s*', ref)
        if a_re:
            author = a_re.group('authors')
        a = []
        if ('et al' in author):
            a = author.split('et al')
            author = a[0]
        temp = []
        if ' and ' in author:
            temp = author.split(' and ')
            author = "".join(temp)
        if ' & ' in author:
            temp = author.split(' & ')
            author = "".join(temp)
        if ' Jr' in author:
            temp = author.split(' Jr')
            author = "".join(temp)
        if ' 2nd' in author:
            temp = author.split(' 2nd')
            author = "".join(temp)
        if ' 3rd' in author:
            temp = author.split(' 3rd')
            author = "".join(temp)
        if ' Sr' in author:
            temp = author.split(' Sr')
            author = "".join(temp)

        all_parsed_authors.append(author)

        # Use counters to see if truth and parsed authors matches
        author_match = True
        parsed_author_string = ""
        for c in author:
            if c.isalnum():
                special_char = "[]&.,-"
                if (c not in special_char) and (c != " "):
                    parsed_author_string += c

        # Strip all special symbols and spaces for both author group string in 
        # ground truth and the parsed authors
        truth_author_arr = row['Author']
        all_truth_authors.append(truth_author_arr)
        truth_author_string = ""
        for word in truth_author_arr:
            for c in word:
                if c.isalnum():
                    special_char = "[]&.,-"
                    if (c not in special_char) and (c != " "):
                        truth_author_string += c

        if Counter(parsed_author_string) != Counter(truth_author_string):
            author_match = False
            # print(author, parsed_author_string, truth_author_string)

        if author_match:
            authors_match.append(1)
            authors_match_count += 1
        else:
            authors_match.append(0)

        # Check title
        if '[' in ref and ']' in ref:     
            temp = re.search('\[(.+?)\]', ref).group(1) 
            if len(temp) > 10 or ' ' in temp:
                title = temp

        all_parsed_titles.append(title)
        all_truth_titles.append(row['Title'])

        # Strip punctuations wrapping around title truth
        if (row['Title'].strip("[].") == title.strip(".")):
            titles_match.append(1)
            titles_match_count += 1
        else:
            titles_match.append(0)


        # Check years
        # Based on the current ground truth, year is wrapped with () but mla style does not require ()
        # Currently set the () a prerequisite for year field
        y = re.search('([\(](?P<year>\d{4})[\)])', ref)
        # Check if the raw ref contains multiple years 
        # e.g. (1946) vol. 101,3 (1976): 88-90. doi:10.1055/s-0028-1104041
        multipleY = re.search('([\(](?P<year>\d{4})[\)])(.+?)([\(]?(\d{4})[\)]?)', ref)
        if y and (not multipleY):
            year = y.group('year').strip()
        elif multipleY:
            year = multipleY.group('year').strip()

        all_parsed_years.append(year)
        all_truth_years.append(row['Year'])

        if (str(row['Year']) == year):
            years_match.append(1)
            years_match_count += 1
        elif (str(row['Year']) == 'YearNotFound') and (year == ""):
            years_match.append(1)
            years_match_count += 1
        else:
            years_match.append(0)


        # Check volumes

        # Check if the issue brackect wrapped both numbers and texts:
        # e.g. Leighton, F., Coloma, L., & Koenig, C. (1975). Structure, composition,
        # physical properties, and turnover of proliferated peroxisomes. 
        # A study of the trophic effects of Su-13437 on rat liver. The Journal of cell biology, 
        # 67(2PT.1), 281–309. https://doi.org/10.1083/jcb.67.2.281
        if ')' in volume and '(' not in volume:
            vi_re = re.search('(?P<volume>[0-9]+)?[\(](?P<issue>([0-9]+\s?[A-Za-z\-\.]+\s?[0-9]+){1,6})[\)]', ref)
            if vi_re:
                issue = vi_re.group('issue')
                volume = vi_re.group('volume')
        elif volume == '':
            vi_re = re.search('(?P<volume>[0-9]+)[\(](?P<issue>[\w\-]+)[\)]', ref)
            i_re = re.search('(?P<volume>[0-9]+)?[\(](?P<issue>[\w\-]{1,3})[\)]', ref)
            # If contains both volume and issue
            if vi_re:
                issue = vi_re.group('issue')
                volume = vi_re.group('volume')
            # If contains no volume
            elif i_re:
                issue = i_re.group('issue')
                if volume == issue:
                    volume = ''
        # Filter the cases when volume and issue were originally identified as volume
        # by the apa_regex_nontext 
        # e.g. 36(1)
        else:
            vi_re = re.search('(?P<volume>[0-9]+)[\(](?P<issue>[\w\-]+)[\)]', volume)
            i_re = re.search('(?P<volume>[0-9]+)?[\(](?P<issue>[\w\-]{1,3})[\)]', volume)
            # If contains both volume and issue
            if vi_re:
                issue = vi_re.group('issue')
                volume = vi_re.group('volume')
            # If contains no volume
            elif i_re:
                issue = i_re.group('issue')
                volume = ''

        all_parsed_volumes.append(volume)            
        all_truth_volumes.append(row['Volume'])

        if (row['Volume'].strip() == volume):
            volumes_match.append(1)
            volumes_match_count += 1
        elif (row['Volume'].strip() == 'VolumeNotFound') and (volume == 'None' or volume == ''):
            volumes_match.append(1)
            volumes_match_count += 1
        else:
            volumes_match.append(0)

        # Check issues
        all_parsed_issues.append(issue)            
        all_truth_issues.append(row['Issue'])

        if (str(row['Issue']).strip() == str(issue).strip()):
            issues_match.append(1)
            issues_match_count += 1
        elif (str(row['Issue']).strip() == 'IssueNotFound') and (str(issue).strip() == 'None' or issue == ''):
            issues_match.append(1)
            issues_match_count += 1
        else:
            issues_match.append(0)
        
        
        # Check publications
        re_pub_one = '[\(](?P<year>(\d{4}))[\)][.]?[ ]?(?P<title>[A-Za-z0-9][\w\:.,?;\%\/\–\-\'\"\+\(\)\[\] ]+[.])\s?(?P<publication>[\w.:\-\;\' ]+)[,]\s([0-9])\s*'
        r_pub_one = re.search(re_pub_one, ref)
        # re_pub_two = '(.\”)\s(?P<publication>[\w.:\-\;\' ]+)\s[(vol.)(,)]\s*'
        # r_pub_two = re.search(re_pub_two, ref)
        if r_pub_one:
            pub = r_pub_one.group('publication').strip()   
        # elif r_pub_two:
        #     pub = r_pub_two.group('publication').strip()

        # Check if the journal name is between '].' and number
        r_foreign_pub = re.search('(].)\s?(?P<publication>[\w\'\-\;:,. ]+)[,]\s[\(]?[0-9][\)]?\s*', ref)
        if r_foreign_pub: 
            pub = r_foreign_pub.group('publication').strip()

        all_truth_publications.append(row['JournalName'])
        all_parsed_publications.append(pub);
        # Strip punctuations wrapping around title truth
        if (row['JournalName'].strip() == pub.strip()):
            publications_match.append(1)
            publications_match_count += 1
        else:
            publications_match.append(0)

        
        # Check pages
        all_parsed_pages.append(page)            
        all_truth_pages.append(row['PageNumber'])
        # truth: 1281-6	original/parsed: 1281–1286
        if('–' in page):
            r_parsedpage = re.search('(?P<one>[0-9]+)[\-\–](?P<two>[0-9]+)s*', page)
            if r_parsedpage:
                parsed_first_page = r_parsedpage.group('one')
                parsed_second_page = r_parsedpage.group('two')
            if(',' in row['PageNumber']):
                r_truepage = re.search('(?P<one>[0-9\-]+)(, )([\w]+(, ))?(?P<two>[0-9\-]+)s*', row['PageNumber'])
                if r_truepage:
                    truth_first_page = r_truepage.group('one')
                    truth_second_page = r_truepage.group('two')
                    truth_second_len = len(truth_second_page)
                    # For ground truth like 15-7, 22-4
                    if('-' in truth_first_page):
                        r_truepage = re.search('(?P<one>[0-9]+)[\-\–](?P<two>[0-9]+)s*', row['PageNumber'])
                        if r_truepage:
                            truth_first_page = r_truepage.group('one')
                    if('-' in truth_second_page):
                        r_truepage = re.search('(?P<one>[0-9]+)[\-\–](?P<two>[0-9]+)s*', row['PageNumber'])
                        if r_truepage:
                            truth_second_page = r_truepage.group('two')
                            truth_second_len = len(truth_second_page)
            elif('-' in row['PageNumber']):
                r_truepage = re.search('(?P<one>[0-9]+)[\-\–](?P<two>[0-9]+)s*', row['PageNumber'])
                if r_truepage:
                    truth_first_page = r_truepage.group('one')
                    truth_second_page = r_truepage.group('two')
                    truth_second_len = len(truth_second_page)
            
            if (truth_first_page == parsed_first_page and truth_second_page == parsed_second_page[-truth_second_len:]):
                pages_match.append(1)
                pages_match_count += 1
            else:
                # print(row['PageNumber'], page, truth_second_page, parsed_second_page, parsed_second_page[-truth_second_len:])
                pages_match.append(0)

        elif (row['PageNumber'].strip() == page.strip()):
            pages_match.append(1)
            pages_match_count += 1
        elif (row['PageNumber'].strip() == 'UNKNOWN') and (page == ''):
            pages_match.append(1)
            pages_match_count += 1
        else:
            pages_match.append(0)


    components = {
        'Raw' : all_raw_strings,
        'Title Truth': all_truth_titles,
        'Title Parsed': all_parsed_titles,
        'Title Match': titles_match,
        'Author Truth': all_truth_authors,
        'Author Parsed': all_parsed_authors,
        'Author Match': authors_match,
        'Volume Truth': all_truth_volumes,
        'Volume Parsed': all_parsed_volumes,
        'Volume Match': volumes_match,
        'Issue Truth': all_truth_issues,
        'Issue Parsed': all_parsed_issues,
        'Issue Match': issues_match,
        'Year Truth': all_truth_years,
        'Year Parsed': all_parsed_years,
        'Year Match': years_match,
        'JournalName Truth': all_truth_publications,
        'JournalName Parsed': all_parsed_publications,
        'JournalName Match': publications_match,
        'PageNumber Truth': all_truth_pages,
        'PageNumber Parsed': all_parsed_pages,
        'PageNumber Match': pages_match
    }

    result = []
    result.append(components)
    result.append(titles_match_count)
    result.append(authors_match_count)
    result.append(volumes_match_count)
    result.append(issues_match_count)
    result.append(years_match_count)
    result.append(publications_match_count)
    result.append(pages_match_count)

    return result
```

## AMA style

**Examples**

Makar AB, McMartin KE, Palese M, Tephly TR. Formate assay in body fluids: application in methanol poisoning. Biochem Med. 1975;13(2):117-126. doi:10.1016/0006-2944(75)90147-7

Smith RJ, Bryant RG. Metal substitutions incarbonic anhydrase: a halide ion probe study. Biochem Biophys Res Commun. 1975;66(4):1281-1286. doi:10.1016/0006-291x(75)90498-2

Old Approach - doesn't work well

```python
ama_regex_end_old = r"(?P<front>.*)(?P<year>\d{4});(?P<volume>[0-9]+)?(?P<issue>\([0-9-\w. ]+\))?:(?P<page>[0-9-–]*)."
```

New Approach - parse first half and second half separately

```python
ama_regex_author = r"(?P<num>[0-9\[\]]+\.?)?\s?(?P<author>(([\w.'-]+[ ]?){1,3}[,][ ])*([\w' -]+)\.(\s*[A-Z]\.)?)(?P<end>.*)"
ama_regex_end = r"(?P<front>.*)(?P<year>\d{4});(?P<volume>[0-9\w.? ]+)?(?P<issue>\([0-9-\w. ]+\))?:?(?P<page>[0-9-–P]*)."
```

```python
import string
from collections import Counter

def testAmaFields():
    ref_style = 'ama'
    raw_ref = []

    authors_truth = []
    authors_parsed = []
    authors_match = []

    journal_truth = []
    journal_parsed = []
    journal_match = []

    title_truth = []
    title_parsed = []
    title_match = []

    year_truth = []
    year_parsed = []
    year_match = []

    volume_truth = []
    volume_parsed = []
    volume_match = []

    issue_truth = []
    issue_parsed = []
    issue_match = []

    page_truth = []
    page_parsed = []
    page_match = []

    for index, row in original_df.iterrows():
        ref = row[ref_style]
        ref_style.lower()
        # Using corresponding style parser
        if ref_style == "ama":
            end_result = re.search(ama_regex_end, ref)
        
        # Check if regex was able to parse the reference
        if end_result:
            raw_ref.append(ref)
            parsed_fields = end_result.groupdict()
        else:
            parsed_fields = False
        
        # Matching
        if parsed_fields:
            ref_front = parsed_fields.get('front')
            # parse the front half
            front_half_result = re.search(ama_regex_author, ref_front)

            author = ""
            title = ""
            journal = ""
            
            if front_half_result:
                author = front_half_result.group('author')
                title_journal = front_half_result.group('end').split('. ')      
                
                if len(title_journal) > 3:
                    for i in range(len(title_journal) - 1):
                        if i != len(title_journal) - 2:
                            title = title + ". " + title_journal[i]
                        else:
                            journal = title_journal[i]
                elif len(title_journal) > 2:
                    title = title_journal[0]
                    journal = title_journal[1]
                else:
                    title = title_journal[0]
            
            
            # Check title
            if title and '[' in title and ']' in title:     
                temp = re.search('\[(.+?)\]', title).group(1)
                if len(temp) > 10 or ' ' in temp:
                    title = temp

            title_parsed.append(title)
            title_truth.append(row['Title'])

            # # # Strip punctuations wrapping around title truth
            if not title:
                title_match.append(0)
            elif (row['Title'].strip("[]. ") == title.strip(". ")):
                title_match.append(1)
            else:
                title_match.append(0)


            # Check journal
            journal_parsed.append(journal)
            journal_truth.append(row['JournalName'])

            if not journal:
                journal_match.append(0)
            elif (row['JournalName'].strip("[]. ") == journal.strip(". ")):
                journal_match.append(1)
            else:
                # Check truncated
                truthList = row['JournalName'].split()
                while 'and' in truthList:
                    truthList.remove('and')
                while 'the' in truthList:
                    truthList.remove('the')
                while 'The' in truthList:
                    truthList.remove('The')
                while 'of' in truthList:
                    truthList.remove('of')
                while 'La' in truthList:
                    truthList.remove('La')
                while 'et' in truthList:
                    truthList.remove('et')
                while 'i' in truthList:
                    truthList.remove('i')
                while 'in' in truthList:
                    truthList.remove('in')
                
                # Get rid of everything after ':'
                if ':' in truthList:
                    tempList = []
                    index = truthList.index(':')

                    for i in range(index):
                        tempList.append(truthList[i])
                    truthList = tempList

                parsedList = journal.split()

                truncatedSituation = False
                validTruncation = True
                if len(truthList) >= len(parsedList):
                    for i in range(len(parsedList)):
                        if len(truthList[i]) != len(parsedList[i]):
                            truncatedSituation = True
                            break

                        if truncatedSituation:
                            for i in range(len(parsedList)):
                                if parsedList[i].lower() not in truthList[i].lower():
                                    validTruncation = False

                if truncatedSituation and validTruncation:
                    journal_match.append(1)
                else:
                    journal_match.append(0)


            # Check authors
            # if ('et al' in author):
            #     a = author.split('et al')
            #     author = a[0]

            temp = []
            if ' and ' in author:
                temp = author.split(' and ')
                author = "".join(temp)
            if ' & ' in author:
                temp = author.split(' & ')
                author = "".join(temp)
            if ' Jr' in author:
                temp = author.split(' Jr')
                author = "".join(temp)
            if ' 2nd' in author:
                temp = author.split(' 2nd')
                author = "".join(temp)
            if ' 3rd' in author:
                temp = author.split(' 3rd')
                author = "".join(temp)
            if ' Sr' in author:
                temp = author.split(' Sr')
                author = "".join(temp)

            authors_truth.append(row['Author'])
            authors_parsed.append(author)

            # Use counters to see if truth and parsed authors matches
            author_match = True
            parsed_author_string = ""
            for c in author:
                if c.isalnum():
                    special_char = "[]&.,-"
                    if (c not in special_char) and (c != " "):
                        parsed_author_string += c

            # Strip all special symbols and spaces for both author group string in 
            # ground truth and the parsed authors
            truth_author_string = ""
            for word in row['Author']:
                for c in word:
                    if c.isalnum():
                        special_char = "[]&.,-"
                        if (c not in special_char) and (c != " "):
                            truth_author_string += c

            if Counter(parsed_author_string) != Counter(truth_author_string):
                author_match = False

            if author_match:
                authors_match.append(1)
            else:
                authors_match.append(0)

            # Check year
            year_truth.append(row['Year'])
            year_parsed.append(parsed_fields.get('year'))

            # Covert to int then compare
            if int(row['Year']) == int(parsed_fields.get('year')):
                year_match.append(1)
            else:
                year_match.append(0)
            
            # Check volume
            volume_truth.append(row['Volume'])
            volume_parsed.append(parsed_fields.get('volume'))

            if (row['Volume'] == parsed_fields.get('volume')):
                volume_match.append(1)
            elif (row['Volume'] == "VolumeNotFound") and (parsed_fields.get('volume') == None):
                volume_match.append(1)
            else:
                volume_match.append(0)
            
            # Check Issue
            issue_truth.append(row['Issue'])
            issue_parsed.append(parsed_fields.get('issue'))

            if parsed_fields.get('issue') == None:
                if row['Issue'] == "IssueNotFound":
                    issue_match.append(1)
                else:
                    issue_match.append(0)
            elif (row['Issue'].translate(str.maketrans('', '', string.punctuation))
             == parsed_fields.get('issue').translate(str.maketrans('', '', string.punctuation))):
                issue_match.append(1)
            else:
                issue_match.append(0)
            
            # Check page
            page = parsed_fields.get('page')
            page_truth.append(row['PageNumber'])
            page_parsed.append(page)


            if page == None:
                page_match.append(0)
            elif page == row['PageNumber']:
                page_match.append(1)
            elif '-' in page and ('-' in row['PageNumber'] or ',' in row['PageNumber']):
                parsedPageList = page.split('-')

                if '-' in row['PageNumber']:
                    truthPageList = row['PageNumber'].split('-')
                elif ',' in row['PageNumber']:
                    truthPageList = row['PageNumber'].split(',')
                
                if parsedPageList[0] in truthPageList[0]:
                    page_match.append(1)
                else:
                    page_match.append(0)
            else:
                page_match.append(0)

    
    components = {
        'Raw Ref': raw_ref,
        'Author Truth': authors_truth,
        'Author Parsed': authors_parsed,
        'Author Match': authors_match,
        'Title Truth': title_truth,
        'Title Parsed': title_parsed,
        'Title Match': title_match,
        'Journal Truth': journal_truth,
        'Journal Parsed': journal_parsed,
        'Journal Match': journal_match,
        'Year Truth': year_truth,
        'Year Parsed': year_parsed,
        'Year Match': year_match,
        'Volume Truth': volume_truth,
        'Volume Parsed': volume_parsed,
        'Volume Match': volume_match,
        'Issue Truth': issue_truth,
        'Issue Parsed': issue_parsed,
        'Issue Match': issue_match,
        'Page Truth': page_truth,
        'Page Parsed': page_parsed,
        'Page Match': page_match
    }

    return components
```
